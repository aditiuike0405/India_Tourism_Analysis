```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

- Statistics of International Tourists to India (2001-2019)

```python
#Fetching data from the csv file

fta_nri_rta=pd.read_csv("IT-2001-2019-nri-ita-fta.csv")
fta_nri_rta.head()
```

```python
#Renaming the data

fta_nri_rta = fta_nri_rta.rename(columns={'FTAs in India (in million)': 'FTA',
                        'NRIs arrivals in India (in million)': 'NRI',
                        'ITAs in India  (in million)': 'ITA'})

fta_nri_rta
```

```python
fta_nri_rta=pd.melt(fta_nri_rta,['Year']).rename(columns={'variable':'Tourists'})
fta_nri_rta
```

```python
#Visualizing the Dataset

plt.figure(figsize=(5,5))
sns.set_context("talk") 
sns.lineplot(x='Year', y='value', hue='Tourists', data=fta_nri_rta)
plt.xlabel("Year")
plt.ylabel("Number of tourists (in millions)")
plt.title("International Tourists To India [2001-2019]")
plt.xticks(range(2001, 2020), rotation=90)
plt.show()
```

- Quaterly Distribution of Tourists (2001-2019)

Total Arrivals

```python
quarterly=pd.read_csv("IT-2001-2019-quaterly.csv")
quarterly.head()
```

```python
#Renaming the dataset

quarterly = quarterly.rename(columns={
    '% Distribution by Quarter - 1st Quarter (Jan-Mar)': '1st Quarter (Jan-Mar)',
    '% Distribution by Quarter - 2nd Quarter(Apr-June)': '2nd Quarter (Apr-June)',
    '% Distribution by Quarter - 3rd Quarter (July-Sep)': '3rd Quarter (July-Sep)',
    '% Distribution by Quarter - 4th Quarter (Oct-Dec)': '4th Quarter (Oct-Dec)'
})
quarterly.head()
```

```python
total_arrivals_yearly = quarterly.groupby('Year')['Arrivals'].sum().reset_index()
total_arrivals_yearly
```

```python
#Visualizing the Dataset

plt.figure(figsize=(7, 7))
sns.barplot(x='Year', y='Arrivals', data=total_arrivals_yearly, palette='viridis')

plt.title("Total Arrivals [2001-2019]", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Arrivals (in million)", fontsize=16)
plt.xticks(fontsize=12,rotation=90)

plt.show()
```

- Quaterly Distribution of FTAs during 2017,2018,2019

```python
#Preparing the Dataset 

quarterly2017=quarterly[quarterly['Year']==2017].copy()
quarterly2017=quarterly2017.drop(['Arrivals','Year'],axis=1)

quarterly2018=quarterly[quarterly['Year']==2018].copy()
quarterly2018=quarterly2018.drop(['Arrivals','Year'],axis=1)

quarterly2019=quarterly[quarterly['Year']==2019].copy()
quarterly2019=quarterly2019.drop(['Arrivals','Year'],axis=1)
```

```python
quarterly2017.head()
```

```python
quarterly2018.head()
```

```python
quarterly2019.head()
```

```python
quarterly_data = {
    '2017': quarterly2017.loc[16,:].values.tolist(),
    '2018': quarterly2018.loc[17,:].values.tolist(),
    '2019': quarterly2019.loc[18,:].values.tolist()
}

# Labels for the quarters
quarter_labels = ['1st Quarter\n(Jan-Mar)', '2nd Quarter\n(Apr-Jun)', '3rd Quarter\n(Jul-Sep)', '4th Quarter\n(Oct-Dec)']

# Colors for the pie chart
colors = ['#a6cee3', '#b2df8a', '#ffff99', '#cab2d6']

# Plotting subplots (pie charts)
fig, axes = plt.subplots(1, 3, sharex=True, figsize=(30, 8))
fig.suptitle('Quarterly Distribution of Tourists [2017, 2018, 2019]', fontsize=30, fontweight='bold')

for ax, (year, quarterly_percentages) in zip(axes, quarterly_data.items()):
    ax.pie(quarterly_percentages, labels=quarter_labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
    ax.set_title(year, fontsize=25, fontweight='bold')

plt.show()
```

```python
total_arrivals_quarterly = quarterly[['1st Quarter (Jan-Mar)', '2nd Quarter (Apr-June)', '3rd Quarter (July-Sep)', '4th Quarter (Oct-Dec)']].sum()
total_arrivals_all_quarters = total_arrivals_quarterly.sum()

# Calculate the average percentage distribution for each quarter
average_percentage_distribution = (total_arrivals_quarterly / total_arrivals_all_quarters) * 100


print("Average Percentage Distribution of Arrivals Quarterly:")
print(average_percentage_distribution)

colors = ['#a7c957', '#80ced7', '#f0a6ca', '#f4d35e']

# Plot pie chart for average percentage distribution
plt.figure(figsize=(5, 5))
plt.pie(average_percentage_distribution, labels=average_percentage_distribution.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Average Percentage Distribution of Arrivals Quarterly')
plt.show()
```

- Statistics of ITAs to India based on purpose 2019

Statistics of ITAs to India based on purpose 2019 (Region-wise)

```python
country_reason=pd.read_csv("IT-2019-regionandreason.csv")
country_reason
```

```python
# Dropping columns
country_reason.drop(['Arrivals (in numbers)','Country of Nationality'],axis=1,inplace=True)
```

```python
#Converting into no.s and renaming columns
country_reason.iloc[:, 1:] = country_reason.iloc[:, 1:] / 100

new_column_names = {
    'Business and Professional(%)': 'Business and Professional',
    'Leisure Holiday and Recreation(%)': 'Leisure Holiday and Recreation',
    'Medical(%)': 'Medical',
    'Indian Diaspora(%)': 'Indian Diaspora',
    'Others(%)': 'Others'
}
country_reason = country_reason.rename(columns=new_column_names)

country_reason
```

```python
# Find the grand total of each column
grand_total = country_reason.iloc[:, 1:].sum()

grand_total
```

```python
regionwise_totals = country_reason.groupby('Region').sum()

# Displaying the region-wise total count of each column
regionwise_totals
```

```python
custom_palette = sns.color_palette("Paired")

# Plotting the data with custom colors
ax = regionwise_totals.plot(kind='bar', figsize=(12, 6), color=custom_palette)

# Adding labels and title
plt.title('Region-wise reasons for visiting India in 2019')
plt.xlabel('Region')
plt.ylabel('Total Count')

# Rotating x-axis labels for better readability
plt.xticks(rotation=86)

# Customize legend position and size
plt.legend(loc='upper left', bbox_to_anchor=(1,1), fontsize='small')

# Displaying the plot
plt.tight_layout()
plt.show()
```

- Popular monuments during 2019,2020,2021

```python
monuments=pd.read_csv("IT-2021-monuments.csv")
monuments
```

```python
# Sort the monuments DataFrame based on total visitors in 2019-20
monuments_sorted = monuments.sort_values(by=['Domestic-2019-20' , 'Foreign-2019-20'], ascending=False)


top_monuments_2019_20 = monuments_sorted.iloc[1:6]

plt.figure(figsize=(14,8))

# Define color codes
color_domestic = '#b1d647'
color_foreign = '#cbaacb'

# Define the width of each bar
bar_width = 0.35

# Define the index for x-axis
index = range(len(top_monuments_2019_20))

# Plotting the bar graph for domestic visitors
plt.bar(index, top_monuments_2019_20['Domestic-2019-20'], bar_width, label='Domestic', color=color_domestic)

# Plotting the bar graph for foreign visitors
plt.bar([i + bar_width for i in index], top_monuments_2019_20['Foreign-2019-20'], bar_width, label='Foreign', color=color_foreign)

# Adding labels and formatting y-axis to display values in thousands
plt.ylabel('Total Visitors (in thousands)')
plt.xlabel('Monuments')
plt.title('Domestic vs Foreign Visitors for Top 5 Tourist Monuments in India (2019-20)')
plt.xticks([i + bar_width / 2 for i in index], top_monuments_2019_20['Name of the Monument'], rotation=90, ha='right')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
monuments_sorted_2020_21 = monuments.sort_values(by=['Domestic-2020-21', 'Foreign-2020-21'], ascending=False)

top_monuments_2020_21 = monuments_sorted.iloc[1:6]

plt.figure(figsize=(14,8))

# Define color codes
color_domestic = '#b1d647'
color_foreign = '#cbaacb'

# Define the width of each bar
bar_width = 0.35

# Define the index for x-axis
index = range(len(top_monuments_2020_21))

# Plotting the bar graph for domestic visitors
plt.bar(index, top_monuments_2020_21['Domestic-2020-21'], bar_width, label='Domestic', color=color_domestic)

# Plotting the bar graph for foreign visitors
plt.bar([i + bar_width for i in index], top_monuments_2020_21['Foreign-2020-21'], bar_width, label='Foreign', color=color_foreign)

# Adding labels and formatting y-axis to display values in thousands
plt.ylabel('Total Visitors (in thousands)')
plt.xlabel('Monuments')
plt.title('Domestic vs Foreign Visitors for Top 5 Tourist Monuments in India (2020-21)')
plt.xticks([i + bar_width / 2 for i in index], top_monuments_2020_21['Name of the Monument'], rotation=90, ha='right')
plt.legend()

plt.tight_layout()
plt.show()
```

- Position of Indian Tourism around the World

```python
worldvsindia=pd.read_csv("IT-2001-2021-worldvsindia.csv")
worldvsindia
```

```python
#Indian Tourism rank throughout years 

years = worldvsindia['Year']
india_tourists = worldvsindia['India - Number (in million)']
ranks = worldvsindia['Rank of India']

# Plotting the line chart
plt.figure(figsize=(14, 8))
plt.plot(years, india_tourists, color='lightcoral', marker='o', markersize=8)

# Adding labels and title
plt.title('Indian Tourism Rank in the World (2001-2019)', fontsize=18)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Number of Tourists (in millions)', fontsize=20)
plt.xticks(years, rotation=90)
plt.grid(False)

# Labeling markers with ranks
for year, tourists, rank in zip(years, india_tourists, ranks):
    plt.text(year, tourists, f"{rank}", ha='right', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Year-wise Comparison of World and India Tourism

years = worldvsindia['Year']
world_tourists = worldvsindia['World - Number (in million)']
india_tourists = worldvsindia['India - Number (in million)']

# Plotting
plt.figure(figsize=(14, 8))
plt.plot(years, world_tourists, label='World', marker='o')
plt.plot(years, india_tourists, label='India', marker='o')

# Adding labels and title
plt.title('Year-wise Comparison of World and India Tourism', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Tourists (in millions)', fontsize=14)
plt.xticks(years, rotation=90)
plt.legend()


plt.tight_layout()
plt.show()
```

```python
years = worldvsindia['Year']
india_percent_change = worldvsindia['India - % Change']

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(years, india_percent_change, color='orange', marker='o')

# Adding labels and title
plt.title('Percentage Change in India Tourism (2001-2019)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Percentage Change', fontsize=14)
plt.xticks(years, rotation=90)

plt.tight_layout()
plt.show()
```
